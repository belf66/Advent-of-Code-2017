#### Day 19 A Series of Tubes
https://adventofcode.com/2017/day/19

##### Description

Somehow, a network packet got lost and ended up here. It's trying to follow a routing diagram (your puzzle input), but it's confused about where to go.

Its starting point is just off the top of the diagram. Lines (drawn with |, -, and +) show the path it needs to take, starting by going down onto the only line connected to the top of the diagram. It needs to follow this path until it reaches the end (located somewhere within the diagram) and stop there.

Sometimes, the lines cross over each other; in these cases, it needs to continue going the same direction, and only turn left or right when there's no other option. In addition, someone has left letters on the line; these also don't change its direction, but it can use them to keep track of where it's been. For example:
<pre>
     |          
     |  +--+    
     A  |  C    
 F---|----E|--+ 
     |  |  |  D 
     +B-+  +--+ 
</pre>
Given this diagram, the packet needs to take the following path:

Starting at the only line touching the top of the diagram, it must go down, pass through A, and continue onward to the first +.
Travel right, up, and right, passing through B in the process.
Continue down (collecting C), right, and up (collecting D).
Finally, go all the way left through E and stopping at F.
Following the path to the end, the letters it sees on its path are ABCDEF.

The little packet looks up at you, hoping you can help it find the way. <strong>What letters will it see (in the order it would see them) if it follows the path?</strong> (The routing diagram is very wide; make sure you view it without line wrapping.)

Your puzzle answer was HATBMQJYZ.

--- Part Two ---
The packet is curious how many steps it needs to go.

For example, using the same routing diagram from the example above...
<pre>
     |          
     |  +--+    
     A  |  C    
 F---|--|-E---+ 
     |  |  |  D 
     +B-+  +--+ 
</pre>
...the packet would go:

- 6 steps down (including the first line at the top of the diagram).
- 3 steps right.
- 4 steps up.
- 3 steps right.
- 4 steps down.
- 3 steps right.
- 2 steps up.
- 13 steps left (including the F it stops on).
This would result in a total of 38 steps.

<strong>How many steps does the packet need to go?</strong>

```python
def day19a(input_str):
    '''
    Solve 2017 Advent of Code day 19 puzzle as described above
    
    Pretty much just follow the characters until you run off the grid
    '''
    lines = input_str.splitlines()
    alpha_list = []
    steps = 0
    idx_i,idx_j = 0,lines[0].find('|')
    dir_i, dir_j = 1,0
    
    idx_i,idx_j = idx_i+dir_i, idx_j+dir_j
    newch = lines[idx_i][idx_j]
    while True:
        if newch.isalpha():
            alpha_list.append(newch)
        elif newch=='+':
            if dir_i==0:
                dir_j = 0
                dir_i = -1 if lines[idx_i-1][idx_j]=='|' else 1
            else:
                dir_i = 0
                dir_j = -1 if lines[idx_i][idx_j-1]=='-' else 1
        idx_i,idx_j = idx_i+dir_i, idx_j+dir_j
        steps += 1
        try:
            newch = lines[idx_i][idx_j]
        except IndexError:
            print('Sequence was "{}", in {} steps'.format(''.join(alpha_list), steps))
            break
```

##### Description

Suddenly, the GPU contacts you, asking for help. Someone has asked it to simulate too many particles, and it won't be able to finish them all in time to render the next frame at this rate.

It transmits to you a buffer (your puzzle input) listing each particle in order (starting with particle 0, then particle 1, particle 2, and so on). For each particle, it provides the X, Y, and Z coordinates for the particle's position (p), velocity (v), and acceleration (a), each in the format <X,Y,Z>.

Each tick, all particles are updated simultaneously. A particle's properties are updated in the following order:

<ol>
<li>Increase the X velocity by the X acceleration.
<li>Increase the Y velocity by the Y acceleration.
<li>Increase the Z velocity by the Z acceleration.
<li>Increase the X position by the X velocity.
<li>Increase the Y position by the Y velocity.
<li>Increase the Z position by the Z velocity.
</li></ol>
Because of seemingly tenuous rationale involving z-buffering, the GPU would like to know which particle will stay closest to position <0,0,0> in the long term. Measure this using the Manhattan distance, which in this situation is simply the sum of the absolute values of a particle's X, Y, and Z position.

For example, suppose you are only given two particles, both of which stay entirely on the X-axis (for simplicity).
```python
import re
def day20(input_str):
    '''
    Solve the first part of the 2017 Advent of Code day 20 puzzle described above
    
    The particle that stays closest in the long run is the one with the lowest 
    acceleration. If more than one have the same acceleration, then choose the 
    one with the lowest initial velocity. And if more than one have the same 
    initial velocity, choose the one with position closest to the origin. 
    '''
    lines = input_str.splitlines()
    particles = []
    idxmin, a_min, v_min, p_min = 0, 1000000, 0, 0
    minimal = []
    for idx,line in enumerate(lines):
        pva = [int(s) for s in re.findall('[0-9-]+', line)]
        a_tot = sum(abs(x) for x in pva[6:])
        v_tot = sum(abs(x) for x in pva[3:6])
        p_tot = sum(abs(x) for x in pva[:3])
        if (a_tot<a_min) or (a_tot==a_min and v_tot<v_min) or (a_tot==a_min and v_tot==v_min and p_tot<p_min):
            minimal = [(idx, a_tot, v_tot, p_tot, pva[6:], pva[3:6], pva[0:3])]
            a_min,v_min,p_min = a_tot,v_tot,p_tot
        elif a_tot==a_min and v_tot==v_min and p_tot==p_min:
            minimal.append((idx, a_tot, v_tot, p_tot, pva[6:], pva[3:6], pva[0:3]))
    return minimal
    
print('particle remaining closest to origin is {}'.format(day20(day20_input_str)))
```

```python
# def day21(input_str):
lines = day21_input_str.splitlines()
rules = dict()
for line in lines:
    k,v = line.split(' => ')
    rules[k] = v

def transpose(r):
    rows = r.split('/')
    cols = [''.join(c) for c in zip(*rows)]
    return '/'.join(cols)

def fliplr(r):
    rows = r.split('/')
    frows = [row[::-1] for row in rows]
    return '/'.join(frows)

def spin(r):
    return r[::-1]

def trans_spin(r):
    return transpose(r[::-1])

def flipud(r):
    return fliplr(r[::-1])

def rotate_right(r):
    return fliplr(transpose(r))

def rotate_left(r):
    return transpose(fliplr(r))

def eight_ways(r):
    '''
    returns eight results of symmetry operations on 2-d array r as strings
    '''
    return set((r, fliplr(r), flipud(r), spin(r), rotate_left(r), trans_spin(r), transpose(r), rotate_right(r)))

def vjoin(r,s):
    if not r:
        return s
    else:
        return '/'.join(r.split('/')+s.split('/'))

def hjoin(r,s):
    if not r:
        return s
    else:
        return transpose(vjoin(transpose(r), transpose(s)))

def expanded_rules(rules):
    new_rules = dict()
    for rule in rules.keys():
        for mod_rule in eight_ways(rule):
            if mod_rule not in rules.keys():
                new_rules[mod_rule] = rules[rule]
            elif rules[mod_rule] != rules[rule]:
                    print("rules don't match -- not adding {} -- {} and {}".format(
                        mod_rule, rules[mod_rule], rules[rule]))
    return new_rules

print(len(rules))
rules.update(expanded_rules(rules))
print(len(rules))   
```

```python
from timeit import default_timer as timer

def squares(pattern, n_iters=10):
    for loop_num in range(n_iters):
        start = timer()
        dim = pattern.index('/')
        if dim % 2 == 0:
            sq_dim = 2
        else:
            sq_dim = 3
        n_squares = dim//sq_dim
        rows = pattern.split('/')
        new_square = []
        for idx in range(n_squares):
            new_row = []
            for jdx in range(n_squares):
                square = []
                for row in rows[idx*sq_dim:(idx+1)*sq_dim]:
                    square.append(row[jdx*sq_dim:(jdx+1)*sq_dim])
                square = '/'.join(square)
                new_row = hjoin(new_row, rules[square])
            new_square = vjoin(new_square, new_row)
        pattern = new_square
        print('iter {}: {} ({:.4f})'.format(loop_num, pattern.count('#'), timer()-start))
    return pattern
```

```python
def make_set(input_str):
    '''
    Convert input string into set of infected positions
    '''
    input_lines = input_str.splitlines()
    infected_set = set()
    for irow in range(len(input_lines)):
        infected_set.update((irow,idx) for idx,c in enumerate(input_lines[irow]) 
                            if c=='#')
    return infected_set   

def make_dict(input_str):
    '''
    Convert input string to dictionary; key is position, value is infection status
    '''
    input_lines = input_str.splitlines()
    infected_dict = dict()
    for irow in range(len(input_lines)):
        infected_dict.update(((irow,idx),1) for idx,c in enumerate(input_lines[irow]) 
                             if c=='#')
    return infected_dict 

def print_grid(infected_set, rowrange=(), colrange=()):
    '''
    Utility to print infection grid.
    
    (Don't use dots because may be mistaken for ellipsis by doctest)
    '''
    if not rowrange:
        rowrange = min(x[0] for x in infected_set), 1+max(x[0] for x in infected_set)
    if not colrange:
        colrange = min(x[1] for x in infected_set), 1+max(x[1] for x in infected_set)
    nrows = rowrange[1]-rowrange[0]+1
    ncols = colrange[1]-colrange[0]+1
    for irow in range(*rowrange):
        row = ''.join(['#' if (irow,icol) in infected_set else '-' for icol in range(*colrange)])
        print(row)
    return # rowrange, colrange

def move_virus(infected_set, steps=1, pos=(0,0), heading=0):
    '''
    >>> test_str = "..#\\n#..\\n..."
    >>> test_set = make_set(test_str)
    >>> move_virus(test_set, steps=3, pos=(1,1))
    2
    >>> print_grid(test_set)
    #-#
    -#-
    >>> test_set = make_set(test_str)
    >>> move_virus(test_set, steps=70, pos=(1,1))
    41
    >>> print_grid(test_set)
    ---##--
    --#--#-
    -#----#
    #-#---#
    #-#--#-
    ---##--
    '''
    heading_d = {0:(-1,0), 1:(0,1), 2:(1,0), 3:(0,-1)} # NESW
    n_infects = 0
    for _ in range(steps):
        if pos in infected_set:
            heading = (heading+1) % 4
            infected_set.discard(pos)
        else:
            heading = (heading-1) % 4
            infected_set.add(pos)
            n_infects += 1
        pos = tuple(v+delta for v,delta in zip(pos,heading_d[heading]))
    return n_infects

def move_virus_d(infected_dict, steps=1, pos=(0,0), heading=0):
    '''
    >>> test_str = '..#\\n#..\\n...'
    >>> test_dict = make_dict(test_str)
    >>> move_virus_d(test_dict, steps=4, pos=(1,1))
    0
    >>> print(test_dict)
    {(0, 2): 1, (1, 0): 2, (1, 1): 0, (0, 0): 0, (0, -1): 0}
    >>> test_dict = make_dict(test_str)
    >>> n = move_virus_d(test_dict, steps=10000000, pos=(1,1))
    >>> print(n)    
    2511944
    '''
    heading_d = {0:(-1,0), 1:(0,1), 2:(1,0), 3:(0,-1)} # NESW
    n_infects = 0
    for _ in range(steps):
        if pos in infected_dict:
            infect_level = infected_dict[pos]
            if infect_level == 0: # is weakened, no change in direction
                infected_dict[pos] += 1
                n_infects += 1
            elif infect_level == 1: # is infected, turn right
                heading = (heading+1) % 4
                infected_dict[pos] += 1
            else: # is flagged, reverse direction
                heading = (heading+2) % 4
                infected_dict.pop(pos)
        else: # absent means clean, turn left
            heading = (heading-1) % 4
            infected_dict[pos] = 0 
        pos = tuple(v+delta for v,delta in zip(pos,heading_d[heading]))
    return n_infects

def day22(input_str, pos=(1,1), steps=1):
    '''
    Solve 2017 Advent of Code day 22 puzzle part 1
    
    >>> test_str = '..#\\n#..\\n...'
    >>> day22(test_str, pos=(1,1), steps=10000)
    5587
    '''
    day22_set = make_set(input_str)
    n_infections = move_virus(day22_set, steps=steps, pos=pos)
    return n_infections
    
    
def day22b(input_str, pos=(1,1), steps=1):
    '''
    Solve 2017 Advent of Code day 22 puzzle part 2
    
    >>> test_str = '..#\\n#..\\n...'
    >>> day22b(test_str, pos=(1,1), steps=10000000)
    2511944
    '''

    day22_dict = make_dict(input_str)
    n_infections = move_virus_d(day22_dict, steps=steps, pos=pos)
    return n_infections
    
# doctest.run_docstring_examples(move_virus, None, verbose=True, name='move_virus')
# doctest.run_docstring_examples(move_virus_d, None, verbose=True, name='move_virus_d')
# doctest.run_docstring_examples(day22, None, verbose=True, name='day22')
# doctest.run_docstring_examples(day22b, None, verbose=True, name='day22b')
```

```python

class Coprocessor():
    '''
    coproc = Coprocessor()
    Coprocessor commands are:
        set -- coproc.set('a',1) -- set register a to 1
        sub -- coproc.sub('a','b') -- add contents of register b to a
        mul -- coproc.mul('a','b') -- multiply a with contents of register b
        jnz -- coproc.jnz('a',3) -- jump 3 steps if register a is not zero
    Each command returns the increment to the next instruction
      (i.e,. 1 if there is no jump)
    This class counts the number of multiply commands executed (coproc.mulcntr)
    
    '''
    
    def __init__(self):
        self.registers = defaultdict(int)
        self.mulcntr = 0
        
    def value_of(self, arg):
        if arg.isalpha():
            return self.registers[arg]
        else:
            return int(arg)
        
    def set(self, reg, arg):
        self.registers[reg] = self.value_of(arg)
        return 1
        
    def sub(self, reg, arg):
        self.registers[reg] -= self.value_of(arg)
        return 1
        
    def mul(self, reg, arg):
        self.registers[reg] *= self.value_of(arg)
        self.mulcntr += 1
        return 1
        
    def jnz(self, reg, arg):
        if self.value_of(reg) != 0:
            return self.value_of(arg)
        else:
            return 1
```

```python
def day23(input_str):
    '''
    Solve the 2017 Advent of Code day23 puzzle (part 1)
    '''
    steps = day23_input_str.splitlines()
    n_steps = len(steps)
    coproc = Coprocessor()
    idx = 0
    cmd, *r = steps[idx].split()
    while True:
        inc = getattr(coproc,cmd)(*r)
        idx += inc
        if not 0<=idx<n_steps:
            break
        cmd, *r = steps[idx].split()
    return coproc.mulcntr

print('number of multiplies was: {}'.format(day23(day23_input_str)))
```

```python
%%time

def day24(input_str):
    '''
    Solve 2017 Advent of Code day24 puzzle as described above
    
    input_str is list of bridge links (one per line) in the format of 17/22
    
    >>> test_input_str = '0/2\\n2/2\\n2/3\\n3/4\\n3/5\\n0/1\\n10/1\\n9/10'
    >>> 
    >>> day24(test_input_str)
    best path strength: 3, length: 31, path: [[0, 1], [1, 10], [9, 10]]
    longest path strength: 4, length: 19, path: [[0, 2], [2, 2], [2, 3], [3, 5]]
    '''
    
    lines = input_str.splitlines()
    lst = [sorted(line.split('/'),key=int) for line in lines]
    lst = [[int(x) for x in l] for l in lst]
    
    q = deque()
    best_pathval = 0
    best_path = [best_pathval,[]]
    longest_path_length = 0
    longest_path_strength = 0
    longest_path = [longest_path_length, longest_path_strength, []]
    l = lst.copy()
    node = 0
    path = []
    q.append((l,[],node)) # q will contain remaining links and path to get to current pos
    while q:
        l,path,node = q.pop()
        possible_links = [link for link in l if node in link]
        if not possible_links:
            pathval = sum(sum(x) for x in path)
            pathlen = len(path)
            if pathval > best_pathval:
                best_path = pathlen, pathval,path
                best_pathval = pathval
            if pathlen > longest_path_length:
                longest_path_length = pathlen
                longest_path = [pathlen, pathval, path]
            elif pathlen == longest_path_length:
                if pathval > longest_path[1]:
                    longest_path[1] = pathval
                    longest_path[2] = path
            continue
        for link in possible_links:
            new_list = l.copy()
            new_list.remove(link)
            new_node = link[1] if link[0]==node else link[0]
            q.append((new_list,path+[link],new_node))
            
    print('best path strength: {}, length: {}, path: {}'.format(*best_path))
    print('longest path strength: {}, length: {}, path: {}'.format(*longest_path))

# doctest.run_docstring_examples(day24, None, verbose=True, name='day24')

day24(day24_input_str)
```

```python
%%time

class Tape():
    '''
    Class to implement a tape (e.g., for a Turing machine)
    
    4 operations available: 
    Tape.write(v) -- write v in current position on tape
    Tape.read() -- read from current position on tape, returns value of current position
    Tape.move(direc) -- move current position in direc + (1) or - (-1) direction
    Tape.checksum() -- return # of 1's currently on tape
    
    >>> tape = Tape()
    >>> tape.write(1)
    >>> tape.write(2)
    >>> tape.move(-1)
    >>> tape.move(-1)
    >>> tape.write(3)
    >>> tape.move(-1)
    >>> tape.write(4)
    >>> print(tape)
    cur_loc: 0 holds val: 4   tape is: deque([4, 3, 0, 2])
    '''
    
    def __init__(self):
        self._cur_loc = 0
        self.tape = deque([0])
        self.len = len(self.tape)
    
    def read(self):
        return self.tape[self._cur_loc]
    
    def write(self, val):
        self.tape[self._cur_loc] = val
                
    def move(self, direc):
        self._cur_loc += direc
        if self._cur_loc < 0:
            self.tape.appendleft(0)
            self._cur_loc += 1
        elif self._cur_loc >= self.len:
            self.tape.append(0)
        self.len = len(self.tape)
        
    def checksum(self):
        return self.tape.count(1)
        
    def __str__(self):
        return 'cur_loc: {} holds val: {}   tape is: {}'.format(
            tape._cur_loc, tape.tape[tape._cur_loc], tape.tape)

def day25(state_ops, num_iters=1):
    '''
    Solve the 2017 Advent of Code puzzle for day 25
    The puzzle should be described above.
    
    >>> test_state_ops = dict(A0=(1,1,'B'), A1=(0,-1,'B'), B0=(1,-1,'A'), B1=(1,1,'A'))
    >>> day25(test_state_ops, 6)
    3
    '''
    tape = Tape()
    state = 'A'
    for _ in range(num_iters):
        val, direc, state = state_ops[state+str(tape.read())]
        tape.write(val)
        tape.move(direc)
    return tape.checksum()

# state_ops is a dictionary describing operations corrensponding to the 
#   day25 input string (translated by me)
# dictionary keys are state+value, i.e., the current state with the value
#   read from the current position on the tape appended to it
# dictionary values are the tuple (write_val, direc, next_state)
state_ops = dict(A0=(1, 1, 'B'), A1=(0,-1,'B'),
                 B0=(1,-1,'C'), B1=(0,1,'E'), 
                 C0=(1,1,'E'), C1=(0,-1,'D'), 
                 D0=(1,-1,'A'), D1=(1,-1,'A'), 
                 E0=(0,1,'A'), E1=(0,1,'F'),
                 F0=(1,1,'E'), F1=(1,1,'A'))

test_state_ops = dict(A0=(1,1,'B'), A1=(0,-1,'B'), B0=(1,-1,'A'), B1=(1,1,'A'))

# doctest.run_docstring_examples(Tape, None, verbose=True, name='Tape')
# doctest.run_docstring_examples(day25, None, verbose=True, name='day25')

print("checksum after running is: {}\n".format(day25(state_ops, num_iters=12683008)))
```
